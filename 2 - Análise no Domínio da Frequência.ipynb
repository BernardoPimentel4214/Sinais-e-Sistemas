    <h2>
        1 - Aproximação de Funções
    </h2>
    <p style="text-indent: 2em;">
        Existem vários problemas na matemática, na física e na engenharia que nos motivam a aproximar uma função por outra. Alguns tipos de funções são mais facilmente diferenciáveis que outras, ou podem não ser prontamente representadas computacionalmente, por exemplo. Um caso clássico de aproximação de funções é a série de Taylor, usada para aproximar sinais elementares por funções polinomiais. Vamos abordar esse tópico de aproximação por meio de um conjunto de sinais ortogonais, chegando a um resultado satisfatório acerca de sinais e sistemas, álgebra linear e equações diferenciais:
    <p style="text-indent: 2em;">
        Deseja-se de obter um sinal f<sub>N</sub>(t) tal que  f<sub>N</sub>(t) ≈ f(t) quando t ∈ [T<sub>1</sub>, T<sub>2</sub>]. Vamos partir da relação
</div>

$$
f_N(t) = \sum_{n=1}^{N} \alpha_n \varphi_n(t). \;\;\;\;\;\;\; (1)
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <ul>
        <li>N: Dimensão de aproximação</li>
        <li>φ: Funções de base (polinômios na série de Taylor)</li>
        <li>α: Coeficientes de aproximação</li>
    </ul>
        f(t), N, φ<sub>n</sub> e [T<sub>1</sub>, T<sub>2</sub>] são dados, tal que nosso objetivo é determinar os coeficientes α<sub>n</sub>. Um indicador da qualidade da nossa aproximação é a energia de um sinal de erro e(t) = f(t) - f<sub>N</sub>(t):
<div>

$$
E_e 
= \int_{T_1}^{T_2} |e(t)|^2\,dt .
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        Idealmente, essa energia deve ser a menor possível. Nesse contexto, um procedimento de minimização de cálculo vem à mente. De fato, esse é um caminho válido para obtenção de α<sub>n</sub>, porém podemos explorar as propriedades bem definidas de espaços vetoriais para resolver esse problema de maneira menos trabalhosa.
    <p style="text-indent: 2em;">
        Sendo assim, cabe basear nossa análise no espaço de Hilbert <i>L</i><sup>2</sup>, cujos elementos são funções ‘quadrado integráveis’. Grande parte dos sinais de interesse prático são tais que o quadrado de seus módulos são integráveis no intervalo [T<sub>1</sub>, T<sub>2</sub>] e portanto, estão contidos em <i>L</i><sup>2</sup>. Vale ressaltar que <i>L</i><sup>2</sup> é um subespaço do espaço de Hilbert, que é mais geral e definido pela presença de um produto interno e o fato de ser <b>completo</b> (ver <a href="https://fma.if.usp.br/~vbchabu/arquivos/chabu-equacoes_diferenciais-espacos_vetoriais_normados.pdf">'chabu equações diferenciais espaços vetoriais normados'</a> para mais informações). Com efeito, vamos usar a existência do produto interno nesse espaço como objeto principal no nosso desenvolvimento.
    <p style="text-indent: 2em;">
        O produto interno entre duas funções é definido por
<div>

$$
<f,g> = \int_{T_1}^{T_2} f(t)\,g^*(t)dt.
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    Pensando nos sinais como vetores no sentido geométrico, percebe-se que o módulo de e(t) é mínimo quando e(t) é ortogonal às funções de base.

</div>

<center>
<left><img src="imagens\bases2D.png" width="500" height="300" alt="bases 2D" ></left>
<right><img src="imagens\bases3D.png" width="500" height="300" alt="bases 3D" ></right>
</center>

<div style="text-align: right; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 14px;">
    Visualização 2D e 3D do princípio da ortogonalidade.
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    Dois vetores são ortogonais quando seu produto interno é nulo, o que nos guia ao princípio da ortogonalidade: A escolha dos coeficientes α<sub>n</sub> que minimiza a energia de e(t) satisfaz &lte(t), φ<sub>n</sub>&gt = 0.
    <p style="text-indent: 2em;">
    	Substituindo o lado direito da relação de e(t) na condição acima, com f<sub>N</sub> substituido pelo somatório nos leva a
</div>

$$
\begin{bmatrix}
\langle \varphi_1, \varphi_1 \rangle & \langle \varphi_2, \varphi_1 \rangle & \cdots & \langle \varphi_N, \varphi_1 \rangle \\
\langle \varphi_1, \varphi_2 \rangle & \langle \varphi_2, \varphi_2 \rangle & \cdots & \langle \varphi_N, \varphi_2 \rangle \\
\langle \varphi_1, \varphi_3 \rangle & \langle \varphi_2, \varphi_3 \rangle & \cdots & \langle \varphi_N, \varphi_3 \rangle \\
\vdots & \vdots & \ddots & \vdots \\
\langle \varphi_1, \varphi_N \rangle & \langle \varphi_2, \varphi_N \rangle & \cdots & \langle \varphi_N, \varphi_N \rangle
\end{bmatrix}
\begin{bmatrix}
\alpha_1 \\ \alpha_2 \\ \alpha_3 \\ \vdots \\ \alpha_N
\end{bmatrix}
=
\begin{bmatrix}
\langle f, \varphi_1 \rangle \\ \langle f, \varphi_2 \rangle \\ \langle f, \varphi_3 \rangle \\ \vdots \\ \langle f, \varphi_N \rangle
\end{bmatrix}.
$$

<div style="text-align: right; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 14px;">
    Se quiser convencer-se desse fato, comece com N baixo como N = 3, montando então um sistema 3x3.
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    Para facilitar esse sistema, vamos supor que as funções de base são como os vetores unitários de uma base canônica, sendo mutuamente ortogonais entre si e por consequência anulando qualquer termo fora da diagonal principal. Isso simplifica nossa substituição original: &ltf(t), φ<sub>n</sub>&gt - α<sub>n</sub>&ltφ<sub>n</sub>, φ<sub>n</sub>&gt = 0.
    Finalmente, encontramos nossos coeficientes α<sub>n</sub>:
    α<sub>n</sub> = &ltf(t), φ<sub>n</sub>&gt / &ltφ<sub>n</sub>, φ<sub>n</sub>&gt.
    No sentido de que a energia do erro tende a 0 quando N tende ao infinito, a nossa igualdade (1) se torna verdadeira.
    <p style="text-indent: 2em;">
        Uma classe de funções mutuamente ortogonais que atendem às condições desse problema são as exponenciais complexas, as quais compõem a série de Fourier.
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h2>
        2 - Série de Fourier
    </h2>
    <h4>
        2.1 - Série de Fourier Complexa
    </h4>
    <p style="text-indent: 2em;">
        Seja x(t) um sinal periódico de período fundamental T<sub>0</sub>. Então, sua representação na série de Fourier complexa é
</div>

$$
x(t) = \sum_{n=-\infty}^{\infty} D_ne^{jn\omega_0t}, \;\;\;\; (2.1)
$$
$$
\mathrm{onde} \;\;\; \omega_0 = \frac{2\pi}{T_0} \;\;\;e \;\;\;D_n = \frac{1}{T_0} \; \int_{T_0}x(t)e^{-jn\omega_0t}dt.
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        A igualdade (2.1) é verdadeira no sentido que o valor quadrático médio do sinal de erro é nulo, de acordo com o teorema de Riesz–Fischer (pontualmente podem haver diferenças). O coeficiente <i>D</i><sub>0</sub> é o valor médio ou termo DC e os demais coeficientes D<sub>n≠0</sub> são os harmônicos.
    <p style="text-indent: 2em;">
        Partiremos então do exemplo de obter a série de Fourier da 'square wave':
</div>

$$
x(t) = \sum_{n=-\infty}^{\infty}\mathrm{rect}\!\left(\frac{t - nT_0}{\tau}\right)
$$

<center><img src="imagens\trem pulsos.png" width="700" height="300" alt="square wave" ></center>

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Queremos obter os coeficientes Dn. Começando com o termo DC (n = 0):
t, T_0, tau, n, w_0 = sp.symbols('t, T_0, tau, n, omega_0')
# Substituindo n = 0 em Dn, temos que D0 = 1/T_0 vezes a integral de x(t) sobre a janela tau, pois é x(t) é nulo por parte do período:
D_0 = 1/T_0 * sp.integrate(1, tau) # O valor dessa square wave é 1 quando não é nula.
display('Termo DC: ', D_0)
```

```python
# Agora, para n > 0 (harmônicos):
D_n = 1/T_0 * sp.integrate(sp.exp(-sp.I * n * w_0 * t), t, conds='none')
display(D_n)

# Avaliando no intervalo (-tau/2, tau/2):
Val_Dn = sp.Eq(sp.symbols('D_n'), D_n.subs(t, tau/2) - D_n.subs(t, -tau/2))
display(Val_Dn)
```

```python
Val_Dn = sp.simplify(Val_Dn) # As duas exponenciais podem ser simplificadas, virando um seno.
display(Val_Dn)

# Substituindo w_0 = 2pi/T_0:
Val_Dn.subs(w_0, 2 * sp.pi / T_0)
```

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        A função especial sen(t)/t comumente recebe a designação de sinc(t) tal que podemos escrever D<sub>n</sub> como 
</div>

$$
D_n = \frac{\tau}{T_0}\mathrm{sinc}\!\left(\frac{n \pi \tau}{T_0}\right)
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        E portanto, a representação em série de Fourier da square wave é: 
</div>

$$
x(t)
=
\frac{\tau}{T_0} + \sum_{n=-\infty, \; n \neq 0}^{\infty} \frac{\tau}{T_0} 
\mathrm{sinc}\!\left(\frac{n \pi \tau}{T_0}\right)
e^{-jn\frac{2 \pi}{T_0}t}
$$

<div style="text-align: right; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 14px;">
    Atenção! A função sinc não tem uma nomenclatura padrão, então sinc pode se referir a algo diferente em outros textos.
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        <b>Fenômeno Gibbs</b>: Fortalecendo o que foi mencionado anteriormente, mesmo tendendo o número de harmônicos ao infinito, é possível que hajam diferenças pontuais entre o sinal verdadeiro e sua representação em série de Fourier (ainda que a energia do erro seja 0). Esse é o caso para sinais descontínuos, ou sinais que não são 'lisos'. Aliás, próximo ao ponto da descontinuidade, a aproximação pode apresentar uma discrepância notável, ultrapassando ou sendo inferior em amplitude à função original por cerca de 9%. Essa oscilação brusca é conhecida como fenômeno Gibbs e é visível nos pontos extremos do gráfico acima.

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h4>
        2.4 - Tópicos Sobre Série de Fourier (<a href="https://aprender3.unb.br/pluginfile.php/3271978/mod_resource/content/6/aula08.pdf">slides aula 07</a>)
    </h4>
    <h5>
        2.4.1 - Teorema de Parseval
    </h5>
    <p style="text-indent: 2em;">
        A ortogonalidade das funções de base da série de Fourier garantem um resultado útil, conhecido como teorema de Parseval: a potência de um sinal periódico é igual a soma das potências de suas componentes de Fourier.
</div>

$$
P_x(t) = \sum_{n=-\infty}^{\infty} |D_n|^{2}
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h5>
        2.4.2 - Existência e Convergência: Condições de Dirichlet
    </h5>
    <p style="text-indent: 2em;">
        Para assertar a existência da série de Fourier de um sinal periódico x(t) qualquer, os coeficientes da série devem ser finitos (a<sub>0</sub>, a<sub>n</sub>, b<sub>n</sub>). Isso é garantido quando o sinal em questão é absolutamente integrável em um período. Porém, isso não implica na convergência da série desse sinal. Foi discutido neste notebook sobre a convergência da série no sentido do sinal de erro tender a 0 com N tendendo ao infinito (convergência na média) e foi exibida a convergência uniforme da aproximação no exemplo gráfico da seção 2.2. Acrescentando a esse desenvolvimento, nota-se que a convergência na média ocorre quando a energia de x(t) é finita em um período e que além disso, existe uma forma de verificar a convergência <b>pontual</b> da aproximação:
    <p style="text-indent: 2em;">
    Condições de Dirichlet - devem ser satisfeitas para que a série de Fourier de um sinal x(t) convirja pontualmente:
    <ul>
        <li>x(t) é absolutamente integrável</li>
        <li>Há um número <b>finito</b> de descontinuidades em um período de x(t)</li>
        <li>Há um número <b>finito</b> de máximos e mínimos em um período de x(t)</li>
    </ul>
    <h5>
        2.4.3 - Como as Amplitudes e Fases Moldam o Sinal no Tempo
    </h5>
    <p style="text-indent: 2em;">
        Retomando a square wave da sessão 2.1, vamos usá-la para demonstrar o papel do espectro de amplitude e do espectro de fase na forma da onda. Começando pelo espectro de amplitude, repare como as frequências baixas contornam o formato geral, enquanto as frequências altas delineiam a estrutura fina e mudanças bruscas do sinal.
</div>

<div style="text-align: right; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 14px;">
    No código abaixo, é usada a forma trigonométrica da square wave com T<sub>0</sub> = π e amplitude unitária.
</div>

```python
from scipy import signal

t = np.linspace(-2*np.pi, 2*np.pi, 1000)

DC      = lambda t: t - t + 0.5
harm_1  = lambda t: 0.5 + (2/np.pi) * (np.cos(t)) # termo DC + 1 harmônica
harm_3  = lambda t: 0.5 + (2/np.pi) * (np.cos(t) - (1/3)*np.cos(3*t) + (1/5)*np.cos(5*t)) # termo DC + 3 harmônicas
harm_5  = lambda t: 0.5 + (2/np.pi) * (np.cos(t) - (1/3)*np.cos(3*t) + (1/5)*np.cos(5*t)
                                                 - (1/7)*np.cos(7*t) + (1/9)*np.cos(9*t)) # termo DC + 5 harmônicas
harm_15 = lambda t: 0.5
harm_15 = harm_15(t)
for n in range(15):
    harm_15 += (2/np.pi) * ((-1)**n / (2*n + 1)) * np.cos((2*n + 1) * t)

curvas  = [DC(t), harm_1(t), harm_3(t), harm_5(t), harm_15, 0.5 * signal.square(t + np.pi/2) + 0.5]
titulos = ["Termo DC", "Termo DC + 1 harmônica", "Termo DC + 3 harmônicas",
           "Termo DC + 5 harmônicas", "Termo DC + 15 harmônicas", "Square Wave"]

fig, axs = plt.subplots(6, 1, figsize=(7 ,7), layout='constrained')
for i in range(len(curvas)):
    axs[i].plot(t, curvas[i], color='purple')
    axs[i].set_title(titulos[i])
    axs[i].grid(True)
axs[0].set_ylim(0, 1)

plt.show()
```

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        NOTA: Sinais que necessitam de muitas harmônicas para construir uma boa aproximação, possuem espectros de amplitude que decaem lentamente. “Podemos mostrar que se as primeiras k - 1 derivadas de um sinal periódico x(t) são contínuas e a k-ésima derivada é descontínua, então seu espectro de amplitude decai com a frequência no mínimo tão rapidamente quanto 1/n<sup>k+1</sup> (LATHI, 2007, p. 548).” Com base nisso, podemos determinar que a brusquidão da square wave resulta num decaimento relativamente lento de 1/n, enquanto o espectro de amplitude de uma onda triangular, comparativamente mais suave, decai com taxa de 1/n<sup>2</sup>.
    <p style="text-indent: 2em;">
        E quanto ao papel do espectro de fase? Este é igualmente importante, pois atua expressivamente na construção de sinais descontínuos e que mudam rapidamente. As harmônicas devem ter sinais opostos em lados distintos de uma descontinuidade e isso é garantido pela fase da harmônica nesses pontos. Sem essa contribuição, teríamos uma forma de onda mais dispersa e pouco representativa do sinal verdadeiro.
    <h5>
        2.4.4 -  Resposta de Sistemas LCIT a Entradas Periodicas
    </h5>
    <p style="text-indent: 2em;">
        Munido dos fatos poderosos de que exponenciais complexas são autofunções de sistemas LCIT e que qualquer sinal periódico pode ser decomposto em uma soma ponderada de exponenciais complexas, surge a motivação de juntar esses resultados de análise no tempo e análise de Fourier para escrever a relação de entrada e saída e<sup>jωt</sup> →  H(jω)e<sup>jωt</sup> como
</div>

$$
\sum_{n=-\infty}^{\infty} D_n e^{jn\omega_0 t} \rightarrow
\sum_{n=-\infty}^{\infty} D_n H(jn\omega_0) e^{jn\omega_0 t}.
$$

    <h2>
        3 - Transformada de Fourier
    </h2>
    <h4>
        3.1 - Representação Espectral de Sinais Aperiódicos
    </h4>
    <p style="text-indent: 2em;">
        Tendo explorado o leque de possibilidades que série de Fourier abre para sinais periódicos, seria agradável se pudéssemos estender esse instrumental para sinais que não se repetem no tempo. Então vamos fazer isso! Imagine um sinal x(t) qualquer, com suporte limitado. Agora imagine que ele se repete como na figura abaixo, com período fundamental T<sub>0</sub>.
</div>

<center><img src="imagens\x e xp.png" width="700" height="350" alt="formas" ></center>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        Variando a duração de T<sub>0</sub>, podemos afastar ou aproximar as parcelas não nulas da versão periódica de x(t). Se T<sub>0</sub> tender ao infinito, recuperamos x(t) original. Por hora, substituindo a versão periódica de x(t) na série de Fourier e tomando um período infinito nos limites da integral de D<sub>n</sub>:
</div>

$$
D_n = \frac{1}{T_0} \int^{\infty}_{-\infty} x(t)e^{-jn\omega_0 t} dt.
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <p style="text-indent: 2em;">
        Para visualizar o efeito que T<sub>0</sub> exerce sobre a forma espectral de x(t), vamos definir o sinal X(ω) de tal forma que os coeficientes D<sub>n</sub> sejam 1/T<sub>0</sub> multiplicado pelas amostras de X(ω):
</div>

$$
X(\omega) = \int^{\infty}_{-\infty} x(t)e^{-j\omega t} dt \;\;\; , \;\;\; D_n = \frac{1}{T_0}X(n\omega_0)
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        Variando T<sub>0</sub>, vemos que a magnitude de D<sub>n</sub> decai quando T<sub>0</sub> aumenta (como esperado), mas que em compensação, o espaçamento entre cada amostra diminui. No caso limite em que T<sub>0</sub> tende ao infinito, os espaçamentos tornam-se infinitesimais, como se agora formassem uma função contínua.
</div>

<center><img src="imagens\Dn se aproximando.png" width="693" height="838" alt="formas" ></center>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        Agora, substituindo D<sub>n</sub> ([1/T<sub>0</sub>] X(nω<sub>0</sub>) na série de Fourier, podemos tomar o limite de T<sub>0</sub>→∞ , que fornece, após algumas pequenas manipulações (fica o exercício):
</div>

$$
x(t) = \frac{1}{2\pi} \int^{\infty}_{-\infty}X(\omega)e^{j\omega t}d\omega.
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
        Essa expressão é conhecida como a transformada de Fourier inversa, ou também como a equação de síntese. Formulando apropriadamente:
</div>

$$
Transformada\;de\;Fourier\;Direta\;(Equa\c{c}ão\;de\;Análise):
$$
$$
\mathscr{F}\{ x(t)\} = X(\omega) = \int^{\infty}_{-\infty}x(t)e^{-j\omega t}dt
$$
$$
Transformada\;de\;Fourier\;Inversa\;(Equa\c{c}ão\;de\;Síntese):
$$
$$
\mathscr{F}\{ X(\omega)\} = x(t) = \frac{1}{2\pi} \int^{\infty}_{-\infty}X(\omega)e^{j\omega t}d\omega
$$
$$
\;
$$
$$
x(t) \leftrightarrow X(\omega)
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h4>
        3.2 - Propriedades da Transformada de Fourier
    </h4>
    <p style="text-indent: 2em;">
    	Seja x<sub>n</sub>(t) um sinal no tempo tal que x<sub>n</sub>(t) ⇔ X<sub>n</sub>(ω) e a<sub>n</sub> uma constante real. Resumidamente, as propriedades da transformada de Fourier são:
        <ul>
            <li>Linearidade:&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; a<sub>1</sub>x<sub>1</sub>(t) + a<sub>2</sub>x<sub>2</sub>(t) ⇔ a<sub>1</sub>X<sub>1</sub>(ω) + a<sub>2</sub>X<sub>2</sub>(ω).</li>
            <li>Deslocamento no Tempo:	&ensp;&emsp;&emsp;&emsp;&emsp; x(t − t<sub>0</sub>) ⇔ e<sup>−jωt<sub>0</sub></sup>X(ω), t<sub>0</sub> > 0.</li>
            <li>Deslocamento na Frequência:	&nbsp;&ensp;&emsp;&emsp; e<sup>jω<sub>0</sub>t</sup>x(t) ⇔ X(ω − ω<sub>0</sub>)</li>
            <li>Diferenciação no Tempo:&emsp;&emsp;&emsp;&emsp;&emsp; d<sup>n</sup>x/dt<sup>n</sup> ⇔ (jω)<sup>n</sup>X(ω).</li>
            <li>Integração no Tempo:&nbsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; ∫<sub>-∞</sub><sup>t</sup>  x(𝜏)d𝜏 ⇔ X(ω)/jω + πX(0)δ(ω).</li>
            <li>Escalonamento no Tempo:&nbsp;&emsp;&emsp;&emsp;&emsp; x(at) ⇔ (1/|a|) X(ω/a)</li>
            <li>Convolução no Tempo:&nbsp;&ensp;&emsp;&emsp;&emsp;&emsp;&emsp; x<sub>1</sub>(t) ∗ x<sub>2</sub>(t) ⇔ X<sub>1</sub>(ω)X<sub>2</sub>(ω)</li>
            <li>Convolução na Frequência:&emsp;&emsp;&emsp;&emsp; x<sub>1</sub>(t)x<sub>2</sub>(t) ⇔ (1/2π) X<sub>1</sub>(ω) * X<sub>2</sub>(ω)</li>
            <li>Conjugado:&nbsp;&ensp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; x<sup>*</sup>(t) ⇔ X<sup>*</sup>(-ω).</li>
            <li>Dualidade:&nbsp;&nbsp;&ensp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; X(t) ⇔ 2πx(-ω).</li>
        </ul>
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h4>
        3.3 - Resposta de Sistemass LCIT no Domínio da Frequência
    </h4>
    <p style="text-indent: 2em;">
    	Sendo um sistema:
        <ul>
            <li>Linear, contínuo e invariante no tempo;</li>
            <li>Estável assintoticamente e quanto a BIBO (podendo essa estabilidade assintotica ser marginal, dado que a entrada do sistema não cause ressonância);</li>
            <li>Caracterizado por entradas x(t) e uma função de resposta ao impulso h(t) que possuem representação no domínio de Fourier, X(ω) e H(ω) respectivamente;</li>
        </ul>
    <p style="text-indent: 2em;">
        podemos, em virtude da propriedade da convolução no tempo, relacionar a entrada à saída de tal sistema por
</div>

$$
Y(\omega) = H(\omega)X(\omega).
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h4>
        3.4 - Energia de Sinais
    </h4>
    <p style="text-indent: 2em;">
    	Para relacionar a energia E<sub>x</sub> de um sinal com sua forma espectral:
</div>

$$
E_x = \int^\infty_{-\infty} |x(t)|^2dt = \int^\infty_{-\infty}x(t)x^*(t)dt \;\Rightarrow\; 
E_x = \int^\infty_{-\infty}x(t)[\frac{1}{2\pi}\int^\infty_{-\infty}X^*(\omega)e^{-j\omega t}d\omega]dt
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
<p style="text-indent: 4em;">
    Trocando a ordem da integração:
</div>

$$
E_x = \frac{1}{2\pi}\int^\infty_{-\infty}X^*(\omega)[\int^\infty_{-\infty}x(t)e^{-j\omega t}dt]d\omega =
\frac{1}{2\pi}\int^\infty_{-\infty}X(\omega)X^*(\omega)d\omega
$$


<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <p style="text-indent: 4em;">
    Finalmente:
</div>

$$
E_x = \int^\infty_{-\infty} |x(t)|^2dt = \frac{1}{2\pi}\int^\infty_{-\infty}|X(\omega)|^2d\omega.
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <p style="text-indent: 2em;">
        Análoga à expressão de 2.4.1, a igualdade acima é conhecida como o teorema de Plancherel-Parseval. Utilizando a mesma linha de operações no produto interno:
</div>

$$
\langle x,y\rangle = \int^\infty_{-\infty}x(t)y^*(t)dt = \frac{1}{2\pi}\int^\infty_{-\infty}X(\omega)Y^*(\omega)d\omega.
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    A relação de Rayleigh (expressão anterior), enuncia que em espaços com produto interno, a transformada de Fourier é um operador unitário (preserva a norma). Em sinais e sistemas, isso garante que a energia do sinal não é alterada pela transformada.
    <p style="text-indent: 2em;">
        Sabemos fisicamente, que todo sinal antrópico deve ter duração finita. Aliás, nenhum sinal do mundo real prolonga-se infinitamente, desde antes do princípio do universo até um futuro impossivelmente distante. Isso significa que todo sinal com qual nós lidamos, possui um suporte ilimitado no domínio de Fourier, mas com uma amplitude que se aproxima de 0 quando ω cresce. Portanto, a energia dos sinais concentra-se em faixas de frequências, de tal modo que podemos considerar apenas as contribuições dessas faixas sem que a energia truncada seja muito diferente da energia verdadeira. Esse é o conceito de <b>largura de banda</b>.
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <h2>
        4 - Simulação Computacional 
    </h2>
    <h4>
        4.1 - Modulação em Amplitude DSB-SC (Telecomunicações)
    </h4>
    <p style="text-indent: 2em;">
        Compilando os conhecimentos adquiridos sobre a transformada de Fourier, vejamos um exemplo de aplicação deles na transmissão de sinais por modulação DSB-SC (<b>Double-Sideband, Suppressed-Carrier</b>):
    <p style="text-indent: 2em;">
        Em um sistema de comunicação, visando evitar interferência entre sinais de rádio que compartilham a mesma estação receptora, ou assimilar o comprimento de onda de um sinal ao tamanho elétrico das antenas do sistema, pela propriedade de deslocamento na frequência, podemos modular um sinal no tempo para atingir esses requisitos. Suponha então, que para essa finalidade, um sinal modulante m(t) (onde m(t) ⇔ M(ω)) deve ser deslocado por uma frequência dada ω<sub>c</sub>. Representando cos(ω<sub>c</sub>t) na forma exponencial complexa, podemos aplicar diretamente a propriedade de deslocamento na frequência, obtendo
</div>


$$
m(t)cos(\omega_c t) \leftrightarrow \frac{1}{2}[M(\omega + \omega_c) + M(\omega - \omega_c)].
$$

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <p style="text-indent: 2em;">
        O efeito que isso exerce sobre o espectro M(ω) é dividi-lo em duas partes, escalonadas pela metade e afastadas simetricamente da origem, com centro em ±ω<sub>c</sub>. As metades dos espectros divididos entre ±ω<sub>c</sub> e a origem são denominadas faixas laterais inferiores, enquanto as outras metades correspondem às faixas laterais superiores (daí vem <b>Double-Sideband</b>, ou seja, faixa lateral dupla). Nos casos em que ω<sub>c</sub> for maior ou igual a 2πB, onde B é a largura de faixa de m(t) em Hz, não ocorrerá sobreposição entre as partes deslocadas do espectro.
    <p style="text-indent: 2em;">
        A fim de recuperar o sinal original (demodulação), similarmente à modulação, podemos dividir o espectro e deslocá-lo por ω<sub>c</sub> para ambos os lados novamente, obtendo o espectro banda-base desejado (porém com a metade da amplitude de M(ω)) e dois espectros colaterais em ±2ω<sub>c</sub>. Finalmente, aplicando um <b>filtro</b> de frequências altas (filtro passa-baixas) no sinal demodulado, podemos eliminar as partes indesejadas em ±2ω<sub>c</sub> e obter o sinal original (salvo a mudança de ½ na amplitude).  Por enquanto, podemos pensar em filtros como o espectro de funções de resposta ao impulso (H(ω)) que atuam sobre o sinal transmitido, pois para que haja uma discussão dedicada sobre eles, necessitamos de ferramentas que veremos no notebook 3. Além disso, como este assunto é relativamente extenso e intrigante, decidi postergá-lo a um notebook dedicado (notebook 4).
</div>

<div style="text-align: left; font-family: 'Cambria Math', 'STIX Math', 'Latin Modern Math', serif; font-size: 20px;">
    <p style="text-indent: 2em;">
        O método de demodulação aqui abordado é o de <b>detecção síncrona</b>, onde a portadora (cos(ω<sub>c</sub>t)) da modulação deve ter a mesma frequência e fase da portadora da demodulação. Nesse contexto, diferenças de fase entre as portadoras podem afetar o espectro de tal forma que o filtro altere o ganho do sinal final, cortando parte de sua energia, enquanto discrepâncias entre as frequências causam efeitos de batimento, oscilando o ganho do sinal demodulado. Quando não essas diferenças não estão presentes, dizemos que a transmissão ocorreu sem distorção (o sinal recuperado é idêntico ao original, a não ser por um múltiplo escalar e por um possível deslocamento no tempo).
</div>

```python
import pimentools as pim
```

```python
w_c = 10                       # Frequência da Onda Portadora
w_m = 1                        # Frequência da Modulante
T = 20                         # Limite do Eixo do Tempo

t = np.linspace(0, T, 1000)
w = np.linspace(-1.5*w_c, 1.5*w_c, 1000)

m = np.cos(w_m*t)
H = 1/(1 + 1j*(w/2))            # Filtro Passa-Baixas de Primeira Ordem Típico

pim.Tx_DSB_SC(m, H, w_c, T, t, w)
```
